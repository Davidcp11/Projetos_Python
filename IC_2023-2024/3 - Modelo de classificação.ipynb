```python
import pandas as pd
import numpy as np
```

```python
Data = pd.read_csv("Dados_relatorios_tratados_unicos.csv")
#Data = Data.sort_values(by="data_formatada", ignore_index=True)
Data["frases_total"] = Data["posi"]+Data["neg"]+Data['neu']
Data = Data.sort_values(by='data_formatada', ignore_index=True)
Data["media_acumulada"] = np.nan
Data["desvio_padrao"] = np.nan
Data["dist_std"] = np.nan
Data
```

```python
# Itere pelas linhas do DataFrame
for i in range(len(Data)):
    # Filtrar as linhas que tÃªm o mesmo valor na coluna "empresas"
    valor_total= []
    for k in range(0, i+1):
        if Data['empresa'].iloc[i] == Data['empresa'].iloc[k]:
            valor_total.append(float(Data['frases_total'].iloc[k]))
    Data["media_acumulada"].iloc[i] = np.mean(valor_total)
    Data["desvio_padrao"].iloc[i] = np.std(valor_total)
    if Data["desvio_padrao"].iloc[i] != np.nan and Data["desvio_padrao"].iloc[i] !=0:
        Data["dist_std"].iloc[i] = (Data["frases_total"].iloc[i] - Data["media_acumulada"].iloc[i])/Data["desvio_padrao"].iloc[i]
    else:
        Data["dist_std"].iloc[i] = np.nan

Data
```

```python
def correlacao(dataframe, empresa):
    new= dataframe[dataframe['empresa'] == empresa]
    new = new.sort_values(by="data_formatada", ignore_index=True)
    new = new.reset_index(drop=True)
    corr = new["retorno_diario_2d"].corr(new["indice"], method='spearman')
    return empresa, corr
```

```python
#Setor financeiro: itau, bradesco, BB, santander = 1
#Setor de energia: petrobras = 2
#Setor de MineraÃ§Ã£o e Metalurgia: vale = 3
#Setor de Bens de Consumo: natura, brf, magazine_luiza = 4
#Setor de Tecnologia: locaweb = 5
#Setor de ConstruÃ§Ã£o e ImobiliÃ¡rio: mrv_engenharia: 6
#Setor de SaÃºde: hapvida = 7
#Setor de Transporte e LogÃ­stica: ccr, gol, azul = 8
'''
    "brf" 2
    "bradesco" 1
    "natura" 4
    "hapvida" 7
    "magazine_luiza" 4
    "santander" 1
    "gol" 8
    "ccr" 8
    "locaweb" 5
    "mrv_engenharia" 6
    "itau" 1
    "vale" 3
    "azul" 8
    "BB" 1
    "petrobras" 2
'''
def atribuir_setor(empresa):
    if empresa in ['bradesco', 'santander', 'itau', 'BB']:
        return 1
    elif empresa in ['brf', 'petrobras']:
        return 2
    elif empresa in ['vale']:
        return 3
    elif empresa in ['magazine_luiza', 'natura']:
        return 4
    elif empresa in ['locaweb']:
        return 5
    elif empresa in ['mrv_engenharia']:
        return 6
    elif empresa in ['hapvida']:
        return 7
    elif empresa in ['ccr', 'gol', 'azul']:
        return 8
    else:
        return None
```

```python
Data["media_acumulada_2d"] = np.nan
Data["desvio_padrao_2d"] = np.nan
Data["dist_std_2d"] = np.nan
Data["dist_std_2d_anterior"] = np.nan
Data["target"] = np.nan
Data["target_anterior"] = np.nan
Data["close_anterior"] = np.nan
for i in range(len(Data)):
    # Filtrar as linhas que tÃªm o mesmo valor na coluna "empresas"
    valor_total= []
    for k in range(0, i):
        if Data['empresa'].iloc[i] == Data['empresa'].iloc[k]:
            valor_total.append(float(Data['retorno_diario_2d'].iloc[k]))
    Data["media_acumulada_2d"].iloc[i] = np.mean(valor_total)
    Data["desvio_padrao_2d"].iloc[i] = np.std(valor_total)
    if Data["desvio_padrao_2d"].iloc[i] != np.nan and Data["desvio_padrao_2d"].iloc[i] !=0:
        Data["dist_std_2d"].iloc[i] = (Data["retorno_diario_2d"].iloc[i] - Data["media_acumulada_2d"].iloc[i])/Data["desvio_padrao_2d"].iloc[i]
    else:
        Data["dist_std_2d"].iloc[i] = np.nan
    if Data["desvio_padrao_2d"].iloc[i] != np.nan:
        if Data["retorno_diario_2d"].iloc[i] < -np.std(valor_total):
            Data["target"].iloc[i] = -1
        elif Data["retorno_diario_2d"].iloc[i] > np.std(valor_total):
            Data["target"].iloc[i] = 1
        else:
            Data["target"].iloc[i] = 0
    else:
        if Data["retorno_diario_2d"].iloc[i] < -0.01:
            Data["target"].iloc[i] = -1
        elif Data["retorno_diario_2d"].iloc[i] > 0.01:
            Data["target"].iloc[i] = 1
        else:
            Data["target"].iloc[i] = 0
```

```python
df1 = df1.dropna()
X = df1[['ind_posi', 'ind_neg', 'ind_neu', 'indice', 'desvio_padrao', 'dist_std', 'setor', 'media_acumulada_2d', 'desvio_padrao_2d',
          'dist_std_2d', 'dist_std_2d_anterior','target_anterior', 'close_anterior', 'Close']]
y = df1['target']

#df1['target'] = df1['retorno_diario_2d'].apply(lambda x: 1 if x > 0.01 else (-1 if x < -0.01 else 0))
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import f_classif
import seaborn as sns
import matplotlib.pyplot as plt


from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,  shuffle=False)


classifiers = {
    'K-Nearest Neighbors': KNeighborsClassifier(n_jobs=-1),
    'SVM': SVC(probability=True, random_state=28),
    'Naive Bayes': GaussianNB(),
    'Random Forest': RandomForestClassifier(n_estimators=1000),
    'Neural Network': MLPClassifier(hidden_layer_sizes=(int((X_train.shape[1] * 2 / 3) + 2), 10),
                                    activation='relu', solver='adam', max_iter=1000, random_state=28)
}

results = {}

for model_name, model in classifiers.items():
    # Treine o modelo no conjunto de treinamento (X_train, y_train)
    model.fit(X_train, y_train)

    # FaÃ§a previsÃµes no conjunto de teste
    y_pred = model.predict(X_test)

    # Calcule a acurÃ¡cia do modelo
    accuracy = accuracy_score(y_test, y_pred)

    # Calcule o relatÃ³rio de classificaÃ§Ã£o do modelo
    report = classification_report(y_test, y_pred, output_dict=True)

    # Calcule a matriz de confusÃ£o
    cm = confusion_matrix(y_test, y_pred)

    # Adapte para o KNeighborsClassifier
    if isinstance(model, KNeighborsClassifier):
        # Calcule a pontuaÃ§Ã£o F estatÃ­stica para cada caracterÃ­stica
        f_scores, _ = f_classif(X_train, y_train)
        # Normaliza para obter a importÃ¢ncia relativa
        feature_importance = f_scores / sum(f_scores)
    elif isinstance(model, RandomForestClassifier):
        # Para RandomForestClassifier
        feature_importance = model.feature_importances_
    elif isinstance(model, SVC) and model.kernel == 'linear':
        # Para SVC com kernel linear
        feature_importance = model.coef_[0]
    else:
        # Se nÃ£o pudermos obter a importÃ¢ncia diretamente, definimos como None
        feature_importance = None

    results[model_name] = {'accuracy': accuracy, 'classification_report': report, 'confusion_matrix': cm, 'feature_importance': feature_importance}

# Exiba os resultados em formato de tabela e matriz de confusÃ£o
for model_name, result in results.items():
    print(f'Modelo: {model_name}')
    print(f'AcurÃ¡cia: {result["accuracy"]:.2f}')
    print('RelatÃ³rio de classificaÃ§Ã£o:\n', result["classification_report"])

    # Exiba as importÃ¢ncias das caracterÃ­sticas
    if 'feature_importance' in result and result['feature_importance'] is not None:
        for feature, importance in zip(X.columns, result['feature_importance']):
            print(f'{feature}: {importance}')
    elif 'feature_importance' in result:
        print('ImportÃ¢ncia das caracterÃ­sticas nÃ£o disponÃ­vel para este modelo.')

    # Crie um DataFrame para a matriz de confusÃ£o
    cm_df = pd.DataFrame(result["confusion_matrix"])

    # Exiba a matriz de confusÃ£o
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm_df, annot=True, fmt='g', cmap='Blues')
    plt.title('Matriz de ConfusÃ£o')
    plt.xlabel('Previsto')
    plt.ylabel('Verdadeiro')
    plt.show()
    print('-----------------------------------------')
```
